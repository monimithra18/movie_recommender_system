# Movie Recommender System 

### Item - Based Collaborative Filtering

```python
# Importing the required packages

import pandas as pd
import numpy as np

from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

import matplotlib.pyplot as plt
import seaborn as seb
```

```python
# Importing the Data Sets

movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
# Now we use the .head() command to a dataframe to print the top 5 rows of the Data Set.

movies.head()
```

```python
# Similarly we use the .head() for the ratings Data Set.

ratings.head()
```

```python
# Now we create a new dataframe where each column and row would be unique which makes things easier to work on.

final_dataset = ratings.pivot(index = 'movieId', columns = 'userId', values = 'rating')
final_dataset.head()
```

```python
# Now we replace all the NaN with 0s to make things understandable to the algorithm.

final_dataset.fillna(0, inplace = True)
final_dataset.head()
```

```python
# Now inorder to maintain the credibility of the output, we add a few filters and try to reduce the noise.

# The following are the filters- 
# 1)To qualify a movie, a minimum of 10 users should have voted a movie.
# 2)To qualify a user, a minimum of 50 movies should have voted by the user. 
```

```python
# Collecting and visualising the number of users who voted and the number of movies that were voted.

no_user_voted = ratings.groupby('movieId')['rating'].agg('count')
no_movies_voted = ratings.groupby('userId')['rating'].agg('count')
```

```python
# Visualising the number of users who voted with our threshold of 10 from the filter

f,ax = plt.subplots(1, 1, figsize = (16, 4))

# ratings['rating'].plot(kind='hist')

plt.scatter(no_user_voted.index, no_user_voted,color = 'cyan')
plt.axhline(y = 10, color = 'r')
plt.xlabel('MovieId')
plt.ylabel('No. of users voted')
plt.show()
```

```python
# Making changes to the dataset with the threshold of 10 for the number of users who voted.

final_dataset = final_dataset.loc[no_user_voted[no_user_voted > 10].index, :]
```

```python
# Visualising the number of votes by each user with our threshold of 50 from the filter.

f,ax = plt.subplots(1, 1, figsize = (16, 4))
plt.scatter(no_movies_voted.index, no_movies_voted, color = 'cyan')
plt.axhline(y = 50, color = 'r')
plt.xlabel('UserId')
plt.ylabel('No. of votes by user')
plt.show()
```

```python
# Making changes to the dataset with the threshold of 50 for the number of votes by each user.

final_dataset = final_dataset.loc[:, no_movies_voted[no_movies_voted > 50].index]
final_dataset
```

```python
# Now to reduce the sparsity, we use the csr_matrix function from the scipy library.

csr_data = csr_matrix(final_dataset.values)
final_dataset.reset_index(inplace = True)
```

```python
# Now for the Movie Recommendation Model, we will be using the kNN algorithm to compute similarity with the
# cosine distance metric.
# We will consider n_neighbors as 20.

knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute', n_neighbors = 20, n_jobs = -1)
knn.fit(csr_data)
```

```python
# Now for making the recommendation system, we first check if the imput movie name is available in 
# the database or not. If it's available, we use the recommendation system to find the similar movies
# and sort them based on their similar distance and output the top 10 movies to be recommended.

def get_movie_recommendation(movie_name):
    n_movies_to_reccomend = 10
    movie_list = movies[movies['title'].str.contains(movie_name)]  
    
    if len(movie_list):        
        movie_idx = movie_list.iloc[0]['movieId']
        movie_idx = final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        distances, indices = knn.kneighbors(csr_data[movie_idx], n_neighbors = n_movies_to_reccomend + 1)    
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(), distances.squeeze().tolist())), key=lambda x: x[1])[:0:-1]
        recommend_frame = []
        
        for val in rec_movie_indices:
            movie_idx = final_dataset.iloc[val[0]]['movieId']
            idx = movies[movies['movieId'] == movie_idx].index
            recommend_frame.append({'Title':movies.iloc[idx]['title'].values[0], 'Distance':val[1]})
            
        df = pd.DataFrame(recommend_frame, index=range(1, n_movies_to_reccomend + 1))
        return df
    
    else:
        return " No movies found. Please check your input and try again! "
```

```python
# Finally, we can get the recommended movies.

get_movie_recommendation('Iron Man')
```

```python
# Another Example

get_movie_recommendation('Memento')
```
